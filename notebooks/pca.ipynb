```python
# Prepare inputs
from cairn.io import load_genotype_array, _select_random_elements_sorted
with yaspin("Preparing input matrix..."):
    g = load_genotype_array(
    zarr_base_path = zarr_p,
    region = "CM023248:1000000-20000000",
    df_samples = metadata,
    genotype_var ='call_genotype',
    pos_var = "variant_position",
    sample_query = 'country == "Ethiopia"'
    )

    ac = g.count_alleles() #Count alleles
    flt = (ac.max_allele() == 1) & (ac[:, :2].min(axis=1) > 1) & (ac.is_segregating()) # Remove singletons and multiallelics, select only segregating sites
    gf = g.compress(flt, axis=0) #Apply filter
    gn = _select_random_elements_sorted(gf, 50_000).to_n_alt() # Select random n snps
    # Convert to n alt alleles
    gn = g.to_n_alt()
```
